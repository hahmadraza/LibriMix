```python
import os
import pandas as pd
import random
import shutil
```

```python

root_dir = "/mnt3/svoice_48k_data/data/"
speaker_ids = []
for dir in os.listdir(root_dir):
    dir_path = os.path.join(root_dir, dir)
    if os.path.isdir(dir_path):
        speaker_ids.append(dir)

speaker_ids.remove("train-clean-360")
speaker_ids.remove("test-clean")
train_ids = random.sample(speaker_ids, round(len(speaker_ids)*0.7))
test_ids = [id for id in speaker_ids if id not in train_ids]
```

```python
train_dir = os.path.join(root_dir, "train-clean-360")
test_dir = os.path.join(root_dir, "test-clean")
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
```

```python
for dir in train_ids:
    src_path = os.path.join(root_dir, dir)
    dst_path = os.path.join(train_dir, dir)
    shutil.move(src_path, dst_path)

for dir in test_ids:
    src_path = os.path.join(root_dir, dir)
    dst_path = os.path.join(test_dir, dir)
    shutil.move(src_path, dst_path)
```

```python
train_ids = os.listdir(train_dir)
test_ids = os.listdir(test_dir)
```

```python
dir_md = pd.DataFrame(columns=['speaker_ID', 'sex', 'subset'])
for id in train_ids:
    speaker_ID = id
    sex = 'M'
    subset = "train-clean-360"

    new_row = {'speaker_ID': speaker_ID, 'sex': sex, 'subset': subset}
    
    # Append the new row to the DataFrame
    dir_md = dir_md.append(new_row, ignore_index=True)

for id in test_ids:
    speaker_ID = id
    sex = 'M'
    subset = "test-clean"

    new_row = {'speaker_ID': speaker_ID, 'sex': sex, 'subset': subset}
    
    # Append the new row to the DataFrame
    dir_md = dir_md.append(new_row, ignore_index=True)
```
